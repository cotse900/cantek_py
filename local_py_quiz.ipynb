```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
players_data = pd.read_csv('players.csv')
```

```python
players_data
```

```python
players_data['height']
```

```python
df = players_data
```

```python
df['feet'] = df['height'].apply(lambda x: int(x.split('-')[0]))
df['inches'] = df['height'].apply(lambda x: int(x.split('-')[1]))
df['feet_in_cm'] = df['feet'] * 30.48
df['inches_in_cm'] = df['inches'] * 2.54
df['height_in_cm'] = df['feet_in_cm'] + df['inches_in_cm']
```

```python
df['height_in_cm']
```

```python
plt.scatter(players_data['height_in_cm'], players_data['weight'])

x = players_data['height_in_cm']
y = players_data['weight']
coefficients = np.polyfit(x, y, 1)
line = np.poly1d(coefficients)

plt.xlabel('Height in cm')
plt.ylabel('Weight in lbs')
plt.title('Scatter plot players\' height vs weight')

plt.show()
```

```python
df['feet'] = df['height'].apply(lambda x: int(x.split('-')[0]))
df['inches'] = df['height'].apply(lambda x: int(x.split('-')[1]))
df['height_in_inches'] = df['feet']*12 + df['inches']
```

```python
df['height_in_inches']
```

```python
plt.scatter(players_data['height_in_inches'], players_data['weight'])

x = players_data['height_in_inches']
y = players_data['weight']

plt.xlabel('Height in inches')
plt.ylabel('Weight in lbs')
plt.title('Scatter plot players\' height vs weight')

plt.show()
```

```python
# coefficients and line are basically the best fit line
plt.scatter(players_data['height_in_inches'], players_data['weight'])

x = players_data['height_in_inches']
y = players_data['weight']
coefficients = np.polyfit(x, y, 1)
line = np.poly1d(coefficients)
plt.plot(x, line(x), color='red', label='Line of Best Fit')

plt.xlabel('Height in inches')
plt.ylabel('Weight in lbs')
plt.title('Scatter plot players\' height vs weight')
plt.legend()

plt.show()
```
