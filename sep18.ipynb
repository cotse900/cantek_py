```python
import pandas as pd
from urllib.parse import urlparse
import numpy as np
import nltk.data
from requests_html import HTMLSession

def parse_elements(url,element):
    session = HTMLSession()
    r = session.get(url)
    elements = r.html.find(element)
    return elements

url = 'https://foreignpolicy.com/2023/09/15/nato-sweden-hungary-membership-democracy/'
paragraphs = parse_elements(url, 'p')
links = parse_elements(url, 'a')
```

```python
for p in paragraphs:
  print(p.text)
```
