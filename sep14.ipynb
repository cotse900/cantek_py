```python
# Plotting the scatter plot
plt.scatter(country_data['Ownership.of.banks'], country_data['Economic.Freedom.Summary.Index'])

x = country_data['Ownership.of.banks']
y = country_data['Economic.Freedom.Summary.Index']
coefficients = np.polyfit(x, y, 1)
line = np.poly1d(coefficients)
plt.plot(x, line(x), color='red', label='Line of Best Fit')

# Customize the plot
plt.xlabel('Ownership of Banks')
plt.ylabel('Economic Freedom Index')
plt.title('Scatter Plot Ownership of Banks vs. Economic Freedom Index')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D toolkit

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with three variables
x = country_data['indiv_libs_vdem_owid']
y = country_data['Economic.Freedom.Summary.Index']
z = country_data['Government.consumption']  # Replace with the name of your third variable

ax.scatter(x, y, z)

# Customize the plot
ax.set_xlabel('individual liberties')
ax.set_ylabel('economic freedom')
ax.set_zlabel('government consumption')
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```
